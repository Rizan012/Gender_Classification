```python
def rem_out(data,col):
    q1 = data[col].quantile(0.25)
    q3 = data[col].quantile(0.75)
    iqr = q3 - q1
    ll,ul = q1 - 1.5 * iqr, q3 + 1.5 * iqr
    data = data[(data[col]>ll) & (data[col]<ul)]
    return data
```

```python
data = rem_out(data,"Height")
```

```python
data.shape
```

```python
sns.boxplot(y="Height",data=data)
```

```python
sns.boxplot(y="Weight",data=data)
```

```python
data = rem_out(data,"Weight")
```
